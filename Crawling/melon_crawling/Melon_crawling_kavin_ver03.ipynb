```python
def get_page_num():
    source = driver.page_source
    bs = BeautifulSoup(source,'lxml')
    page_length = []
    page_length = bs.select('#pageObjNavgation > div > span > a')
    len(page_length) #  9면 10페이지인것
    return len(page_length)
```
