```python
import pandas as pd
import numpy as np
```

```python
idol_withid = pd.read_csv('idol_list_idfix.csv', encoding='euc-kr')
idol_withid.head()
```

```python
idol_withid[idol_withid.artist == '신화']
```

```python
# 불필요한 열 드랍
idol_withid = idol_withid.drop(idol_withid.columns[[1, 3]], axis=1)
idol_withid.head(3)
```

```python
# idol_id int로 수정
idol_withid = idol_withid[np.isfinite(idol_withid['idol_id'])].reset_index(inplace=False)
idol_withid['idol_id'] = idol_withid['idol_id'].astype(int)
idol_withid.info()
```

### 소속사 전처리

```python
id1 = idol_withid.copy()
```

```python
id1.agency.unique()
```

엔터테인먼트라는 단어가 굉장히 많고 사실상 SM 엔터테인먼트는 SM으로 부르기 때문에 엔터테인먼트를 제거하기로 결정.

```python
id1.agency = id1.agency.str.replace('엔터테인먼트', '')
id1.agency.unique()
```

```python
# 공백 제거, .제거, 결성은? 제거
id1.agency = id1.agency.str.replace(' ', '')
id1.agency = id1.agency.str.replace('.', '')
id1.agency = id1.agency.str.replace('결성은?', '')
id1.agency = id1.agency.str.replace(',', '')
# id1.agency = id1.agency.str.replace('?(', '')

id1.agency.unique()
```

```python
id1.agency = id1.agency.str.replace('?(', '', regex=False)
id1.agency.unique()
```

```python
# idol_withid.agency.str.split('?', expand=True)
agency_split = id1.agency.str.split('?', expand=True)
agency_split.head()
```

```python
agency_split.columns = ['new', 'old']
agency_split.old = agency_split.old.str.replace(')', '', regex=False)
```

```python
id1['agency_old'] = agency_split['old']
id1['agency_new'] = agency_split['new']
id1 = id1.drop('agency', axis=1)
id1.head()
```

```python
id1 = id1.drop(id1.columns[0], axis=1)
```
