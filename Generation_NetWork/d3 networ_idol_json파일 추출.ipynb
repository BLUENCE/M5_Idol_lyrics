```python
import pandas as pd
import json
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
# 파일 불러오기
lyricist_gen1 = pd.read_csv('idol_generation_1_kavin_ver01.csv', encoding = 'euc-kr')
lyricist_gen2 = pd.read_csv('idol_generation_2_kavin_ver01.csv', encoding = 'utf-8')
lyricist_gen3 = pd.read_csv('idol_generation_3_kavin_ver01.csv', encoding = 'utf-8')
```

```python
# 맨 앞 Unnamed: 0 칼럼 제거
lyricist_gen1 = lyricist_gen1.drop(['Unnamed: 0'], axis = 1)
lyricist_gen2 = lyricist_gen2.drop(['Unnamed: 0'], axis = 1)
lyricist_gen3 = lyricist_gen3.drop(['Unnamed: 0'], axis = 1)
```

```python
# NaN값 제거, lyricist에 있는 NA만을 제거
lyricist_gen1 =  lyricist_gen1.dropna(subset=['lyricist'])
lyricist_gen1 = lyricist_gen1.reset_index()
lyricist_gen2 =  lyricist_gen2.dropna(subset=['lyricist'])
lyricist_gen2 = lyricist_gen2.reset_index()
lyricist_gen3 =  lyricist_gen3.dropna(subset=['lyricist'])
lyricist_gen3 = lyricist_gen3.reset_index()
```

```python
# 맨 앞 Unnamed: 0 칼럼 제거
lyricist_gen1 = lyricist_gen1.drop(['index'], axis = 1)
lyricist_gen2 = lyricist_gen2.drop(['index'], axis = 1)
lyricist_gen3 = lyricist_gen3.drop(['index'], axis = 1)
```

### 1) Gen1작업

```python
def preprocess(text1, text2):
    splitArr = list(re.split(', ', text1))
    artist_name = list(text2)

    lyricist = []
    artist = []
    
    lyricist.clear()
    artist.clear()

    i = 0
    for i in range(0, len(splitArr)):
        lyricist.append(splitArr[i].strip())
        artist.append(text2)
        i = i + 1
    result = [lyricist, artist]
    return result
```

```python
# lyricist_gen1의 가수 목록 데이터프레임 생성
i = 0
a = 0
b = 0
lyricist = []
artist = []

lyricist.clear()
artist.clear()
    
for i in range(0, len(lyricist_gen1)):
    try:
        for a in range(0, len(preprocess(lyricist_gen1.lyricist[i], lyricist_gen1.artist[i])[0])):
            lyricist.append(str(preprocess(lyricist_gen1.lyricist[i], lyricist_gen1.artist[i])[0][a]))
        for b in range(0, len(preprocess(lyricist_gen1.lyricist[i], lyricist_gen1.artist[i])[1])):
            artist.append(str(preprocess(lyricist_gen1.lyricist[i], lyricist_gen1.artist[i])[1][b]))
    except:
        lyricist.append('')
        artist.append('')
    
preprocessing_gen1 = pd.DataFrame({"lyricist" : lyricist, "artist" : artist, "generation" : 1})
preprocessing_gen1
```

```python
preprocessing_gen1.rename(columns={"lyricist":"source","artist":"target"}, inplace=True)
```

```python
grouped_idol_gen1 = preprocessing_gen1.groupby(["source","target"]).size().reset_index()
grouped_idol_gen1.head(20) # 갑자기 빈 행이 존재함 ->삭제
```
