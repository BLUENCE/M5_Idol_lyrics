# 포켓몬 네트워크 시각화

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv('pokemon_data/Pokemon.csv')
pokemon.shape
```

```python
pokemon.head(10)
```

```python
g = nx.Graph()
g=  nx.from_pandas_edgelist(pokemon, source='Name', target='Type 1')
print(nx.info(g))
```

```python
plt.figure(figsize=(20, 20))
pos = nx.spring_layout(g, k=0.15)
nx.draw_networkx(g,pos,node_size=25, node_color='blue')
plt.show()
```

```python
G = nx.Graph()
```

```python
df_edges = pd.read_csv('pokemon_data/stack_network_links.csv')
df_nodes = pd.read_csv('pokemon_data/stack_network_nodes.csv')
print(df_links.shape)
print(df_nodes.shape)
```

```python
plt.figure(figsize=(25, 25))

options = {
    'edge_color': '#FFDEA2',
    'width' : 1,
    'with_labels' : True,
    'font_weight' : 'regular'
}
```

```python
size = [G.node[node]['nodesize']*10 for node in G]
```

```python
color_map = {1:'#f09494', 2:'#eebcbc', 3:'#72bbd0', 4:'#91f0a1', 5:'#629fff', 6:'#bcc2f2',  
             7:'#eebcbc', 8:'#f1f0c0', 9:'#d2ffe7', 10:'#caf3a6', 11:'#ffdf55', 12:'#ef77aa', 
             13:'#d6dcff', 14:'#d2f5f0'} 

colors = [color_map[G.node[node]['group']] for node in G]
```

```python
for index, row in df_nodes.iterrows():
    G.add_node(row['name'], group=row['group'], nodesize=row['nodesize'])

for index, row in df_edges.iterrows():
    G.add_weighted_edges_from([(row['source'], row['target'], row['value'])])

nx.draw(G, node_color=colors, node_size=size, pos=nx.spring_layout(G, k=0.25, iterations=50), **options)
ax = plt.gca()
ax.collections[0].set_edgecolor("#555555")
plt.show()
```
