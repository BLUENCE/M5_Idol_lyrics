```python
import pandas as pd
import re
from pandas import DataFrame
```

```python
d1 = pd.read_csv('song_data_yewon_ver02.csv')
```

```python
d1.info()
```

```python
hangs = d1.creator
```

```python
hangs[0]
```

```python
m = re.search('^(.+) 작사(.+) 작곡(.+) 편곡', str(hangs[0]))
str(m.group(1))
```

여기!

```python
lyricist = []
composer = []
arranger = []

for i in range(0, len(hangs)):
    m = re.match('^(.+) 작사(.+) 작곡(.+) 편곡', str(hangs[i]))
    try:
        l = str(m.group(1))
        lyricist.append(l) 
        c = str(m.group(2))
        composer.append(c)
        a = str(m.group(3))
        arranger.append(a)
    except:
        lyricist.append('nan')
        composer.append('nan')
        arranger.append('nan')
```

```python
d2 = DataFrame({"lyricist": lyricist, "composer":composer, "arranger":arranger})
```
