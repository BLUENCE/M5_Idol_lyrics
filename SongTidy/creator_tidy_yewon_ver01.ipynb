```python
import pandas as pd
import re
from pandas import DataFrame
```

```python
d1 = pd.read_csv('song_data_yewon_ver02.csv')
```

```python
d1.info()
```

```python
hangs = d1.creator
```

```python
hangs[0]
```

```python
m = re.search('^(.+) 작사(.+) 작곡(.+) 편곡', str(hangs[0]))
str(m.group(1))
```

여기!

```python
lyricist = []
composer = []
arranger = []

for i in range(0, len(hangs)):
    m = re.match('^(.+) 작사(.+) 작곡(.+) 편곡', str(hangs[i]))
    try:
        l = str(m.group(1))
        lyricist.append(l) 
        c = str(m.group(2))
        composer.append(c)
        a = str(m.group(3))
        arranger.append(a)
    except:
        lyricist.append('nan')
        composer.append('nan')
        arranger.append('nan')
```

```python
d2 = DataFrame({"lyricist": lyricist, "composer":composer, "arranger":arranger})
```

```python
d2.info()
```

```python
d2.tail(100)
```

```python
d2['lyricist'] = d2['lyricist'].str.replace(r'작사',',')
d2['composer'] = d2['composer'].str.replace(r'작곡',',')
d2['arranger'] = d2['arranger'].str.replace(r'편곡',',')
```

```python
d2.tail(100)
```

```python
# 머지
df1 = pd.merge(d1, d2, how='outer', left_index=True, right_index=True)
```
