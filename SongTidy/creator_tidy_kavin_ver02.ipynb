```python
path = "C:/Users/Kavin/Projects/M5_Idol_lyrics/Data/song_data_yewon_ver02.csv" # 불러올 경로설정
song = pd.read_csv(path, encoding = 'utf-8')
# song = song.drop(columns=['Unnamed: 0'])
```

```python
# 불필요 문자열(\r, \n) 제거
song['creator'] = song['creator'].str.replace('\r', ' ')
song['creator'] = song['creator'].str.replace('\n', ' ')
```

```python
song.head()
```

```python
# 전처리 함수
def preprocess(text):
    splitArr = list(filter(None, re.split("(작사)|(작곡)|(편곡)", text)))

    lyricist = []
    composer = []
    arranger = []
    
    lyricist.clear()
    composer.clear()
    arranger.clear()

    i = 0
    for i in range(0, len(splitArr)):
        if splitArr[i] == "작사":
            lyricist.append(splitArr[i-1].strip())
        elif splitArr[i] == "작곡":
            composer.append(splitArr[i-1].strip())
        elif splitArr[i] == "편곡":
            arranger.append(splitArr[i-1].strip())
        i = i + 1
    result = [', '.join(lyricist), ', '.join(composer), ', '.join(arranger)]
    return result
```

```python
# 행마다 작사/작곡/편곡가 전처리 결과 보기
preprocess(song.creator[0])
```

```python
# song 데이터프레임 전처리함수 이용하여 전처리 후 dataframe 추가로 만들기
i = 0
lyricist = []
composer = []
arranger = []

lyricist.clear()
composer.clear()
arranger.clear()
    
for i in range(0, len(song)):
    try:
        lyricist.append(str(preprocess(song.creator[i])[0]))
        composer.append(str(preprocess(song.creator[i])[1]))
        arranger.append(str(preprocess(song.creator[i])[2]))
    except:
        lyricist.append('')
        composer.append('')
        arranger.append('')
    
preprocessing_result = pd.DataFrame({"lyricist" : lyricist, "composer" : composer, "arranger" : arranger})
```

```python
preprocessing_result.head()
```

```python
# 두 개의 데이터프레임 길이가 같은지 확인
len(song) == len(preprocessing_result)
```

```python
# 두 개의 데이터프레임 합치기
df = pd.concat([song, preprocessing_result], axis=1)
```
