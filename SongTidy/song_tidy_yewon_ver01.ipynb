## 드랍시킬 것들
- 가사가 없는 행
- 가수가 없는 행

## 특수한 경우들 
1. 언어가 한국어가 아닌 노래 제거
- Japanese Chniese English
2. Instrumental 제거
3. Remix 혹은 가사는 동일한 중복곡 제거
4. 프로그램에서 부른 노래 제거
- 불후의 명곡, 나는 가수다 등
- 본인의 곡이 아닌 경우가 많아서 반드시 제거해주어야 함

```python
import pandas as pd
import numpy as np
import re
from plotnine import *
```

```python
df = pd.read_csv('song_data_fixed.csv')
```

```python
df
```

```python
# 인덱스칼럼, 가사가 없는 행, 가수가 없는 행 드랍
df = df.drop(df.columns[0], axis=1)
df = df[pd.notnull(df['lyrics'])]
df = df[pd.notnull(df['artist'])]
```

```python
df
```

```python
df['lyrics'] = df['lyrics'].str.replace(r'\r\r\n','<br>')
df['creator'] = df['creator'].str.replace(r'\r|\n',' ')
```

```python
# is_title이 nan일 때 = '수록곡' 
df['is_title'] = df['is_title'].fillna('수록곡')
```

```python
df
```

```python
# 제목에 리믹스,라이브,inst,영일중,ver 인 행 제거
df2 = df[df.title.str.contains(r'\(.*\s*([Rr]emix|[Mm]ix|[Ll]ive|[Ii]nst|[Cc]hn|[Jj]ap|[Ee]ng|[Vv]er)\s*.*\)') == False]
df2
```

```python
# 앨범에 나가수, 불명, 복면인 행 제거
df3 = df2[df2.album.str.contains(r'(가수다|불후의|복면가왕)') == False]
df3
```

```python
# 타이틀에 히라가나/가타카나가 들어간 행 삭제
df4 = df3[df3.title.str.contains(u'[\u3040-\u309F\u30A0-\u30FF\u31F0-\u31FF]+') == False]
df4
```

```python
# 가사에 히라가나/가타카나가 들어간 행 삭제
df5 = df4[df4.lyrics.str.contains(u'[\u3040-\u309F\u30A0-\u30FF\u31F0-\u31FF]+') == False]
df5
```

```python
# 앨범에 히라가나/가타카나가 들어간 행 삭제
df6 = df5[df5.album.str.contains(u'[\u3040-\u309F\u30A0-\u30FF\u31F0-\u31FF]+') == False]
df6
```

```python
df6
```

```python
df6.to_csv('song_data_yewon_ver01.csv', index=False)
```
