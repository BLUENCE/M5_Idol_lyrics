### song_data_yewon_ver03.csv파일을 아티스트의 중복노래를 제거한 노트북입니다.

```python
import pandas as pd
import re
```

```python
s_df = pd.read_csv('tidydata/song_data_yewon_ver03.csv')
```

```python
s_df.loc[14]
```

```python
s_df.info()
```

```python
# df.drop_duplicates을 사용, 'artist'기준으로 'title'중복값을 제거합니다.
# keep='first'의 의미는 맨윗줄을 기준으로 밑에 있는 중복title행을 날린다는 의미입니다.
test = s_df.drop_duplicates(['artist', 'title'], keep='first')
```

```python
# fakelove를 검색하면 중복값이었던 행들이 제거된 것을 확인할 수 있습니다.
test.loc[test['artist'].str.contains(r'(방탄소년단)', regex=True)]
```

```python
# 약 2000개의 중복값이 제거되었습니다.
test.info()
```

```python
test.to_csv('tidydata/song_data_yewon_ver04.csv', index=False)
```
