```python
import pandas as pd
import re
import numpy as np
import os
import csv
from pandas import DataFrame
```

```python
s_df = pd.read_csv('./project/idol/song_data_yewon_ver05.csv', header=0,  encoding = 'utf-8', low_memory=False)
s_df.head()
```

```python
a_df = pd.read_csv('./project/idol/M5_Idol_lyrics/Data/Data20180915/artist_info_combined_ver01.csv', header=0,  encoding = 'utf-8', low_memory=False)
a_df.head()
```

```python
d = pd.merge(s_df, a_df, on='artist')
```

```python
# 아티스트데이터와 노래데이터 머지
df = d.copy()
df.head()
```

```python
df2 = df[['title', 'artist', 'gender', 'lyrics',  'release_date', 'debut_y']]
df2.head()
```

```python
release = df2.debut_y
```

```python
# release_date으로 debut_year칼럼 새로 만들어주기
m = re.search('^(\d{4})', str(release[0]))
str(m.group(1))
```

```python
r =[]
for i in range(0, len(release)):
    m = re.match('^(\d{4})', str(release[i]))
    try:
        l = str(m.group(1))
        r.append(l)
    except:
        r.append('nan')
```

```python
d = DataFrame({"debut_year": r})
```

```python
d.info()
```

```python
release_df = pd.merge(df2, d, how='outer', left_index=True, right_index=True)
```

```python
# 잘담겼는지 확인
release_df = release_df[['title', 'artist', 'gender', 'lyrics', 'release_date', 'debut_year']]
release_df.head()
```

```python
# 데뷔년도가 없는게...생각보다 많다 ㅠㅠ
release_df.loc[release_df['debut_year'].str.contains(r'(nan)', regex=True)]
```

```python
df = release_df
```
