```python
import pandas as pd
import re
import numpy as np
import os
import csv
from pandas import DataFrame
```

```python
s_df = pd.read_csv('C:/Users/Kavin/Projects/M5_Idol_lyrics/SongTidy/FinalTidy/tidydata/lyricist_tidy03.csv', header=0,  encoding = 'utf-8', low_memory=False)
s_df.head(100)
# len(s_df)
```

```python
a_df = pd.read_csv('C:/Users/Kavin/Projects/M5_Idol_lyrics/Data/Data20180921/artist_info_combined_ver04.csv', header=0,  encoding = 'euc-kr', low_memory=False)
a_df.head(100)
# len(a_df)
```

```python
d = pd.merge(s_df, a_df, on='artist')
```

```python
# 아티스트데이터와 노래데이터 머지
df = d.copy()
df.head()
```

```python
df2 = df[['title', 'artist', 'gender', 'lyricist', 'composer', 'arranger', 'debut_y']]
df2.head()
```

```python
release = df2.debut_y
```

```python
# release_date으로 debut_year칼럼 새로 만들어주기
m = re.search('^(\d{4})', str(release[0]))
str(m.group(1))
```

```python
r =[]
for i in range(0, len(release)):
    m = re.match('^(\d{4})', str(release[i]))
    try:
        l = str(m.group(1))
        r.append(l)
    except:
        r.append('nan')
```

```python
d = DataFrame({"debut_year": r})
d.head()
```

```python
d.info()
```

```python
release_df = pd.merge(df2, d, how='outer', left_index=True, right_index=True)
```

```python
# 잘담겼는지 확인
release_df = release_df[['title', 'artist', 'gender', 'lyricist', 'composer', 'arranger', 'debut_year']]
release_df.head()
```

```python
# 데뷔년도가 없는게...생각보다 많다 ㅠㅠ
release_df.loc[release_df['debut_year'].str.contains(r'(nan)', regex=True)]
```

```python
df = release_df
```

```python
# 0세대~3세대로 구분
df0 = df[df["debut_year"].str.contains('1992|1993|1994|1995', na=False)]
df1 = df[df["debut_year"].str.contains('1996|1997|1998|1999|2000|2001|2002|2003', na=False)]
df2 = df[df["debut_year"].str.contains('2004|2005|2006|2007|2008|2009|2010|2011', na=False)]
df3 = df[df["debut_year"].str.contains('2012|2013|2014|2015|2016|2017|2018', na=False)]
```

```python
df0["generation"] = 0
df1["generation"] = 1
df2["generation"] = 2
df3["generation"] = 3
```

```python
# 1세대 2세대 3세대 csv파일로 저장
df1.to_csv('C:/Users/Kavin/Projects/study/idol_generation_1_kavin_ver01.csv', encoding='')
df2.to_excel('C:/Users/Kavin/Projects/study/idol_generation_2_kavin_ver01.xlsx', sheet_name='sheet1')
df3.to_excel('C:/Users/Kavin/Projects/study/idol_generation_3_kavin_ver01.xlsx', sheet_name='sheet1')
```
