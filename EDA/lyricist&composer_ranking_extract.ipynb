```python
def preprocess(text1, text2):
    splitArr = list(re.split(', ', text1))
    artist_name = list(text2)

    lyricist = []
    artist = []
    
    lyricist.clear()
    artist.clear()

    i = 0
    for i in range(0, len(splitArr)):
        lyricist.append(splitArr[i].strip())
        artist.append(text2)
        i = i + 1
    result = [lyricist, artist]
    return result
```

## (2) 데이터 전처리 - 작곡가 칼럼 전부 분리하기

```python
idol_co.loc[idol_co['composer'] == "na"]
```

```python
idol2 = idol_co[["artist","composer"]]
```

```python
# song 데이터프레임 전처리함수 이용하여 전처리 후 dataframe 추가로 만들기
i = 0
a = 0
b = 0
composer = []
artist = []

composer.clear()
artist.clear()
    
for i in range(0, len(idol2)):
    try:
        for a in range(0, len(preprocess(idol2.composer[i], idol2.artist[i])[0])):
            composer.append(str(preprocess(idol2.composer[i], idol2.artist[i])[0][a]))
        for b in range(0, len(preprocess(idol2.composer[i], idol2.artist[i])[1])):
            artist.append(str(preprocess(idol2.composer[i], idol2.artist[i])[1][b]))
    except:
        composer.append('')
        artist.append('')
    
preprocessing_result2 = pd.DataFrame({"composer" : composer, "artist" : artist})
```
